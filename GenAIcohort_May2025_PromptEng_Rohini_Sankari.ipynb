```python
def get_response(prompt):
  #define model
  model="groq/llama3-70b-8192"
  messages = [{"role": "user", "content": prompt}]

  #call responses
  response = completion(
      model=model,
      messages=messages
  )

  return response['choices'][0]['message']['content'] or ""





def clean_text(text):
    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)

    # Remove punctuation and digits
    text = re.sub(r'[^a-z\s]', '', text)



    return text

```

## Conversation Summary

Summarizes the conversation and stores the current summary in memory.

Explore it yourself!

```python
!pip install langchain_groq
```

```python
from langchain.chains import ConversationChain
from langchain.memory import ConversationSummaryMemory
from langchain_groq import ChatGroq

llm = ChatGroq(
    groq_api_key=os.environ["GROQ_API_KEY"] ,
    model_name="llama3-70b-8192"
)
# 2. Create a summarizing memory
summary_memory = ConversationSummaryMemory(
    llm=llm,
    return_messages=True  # optional: to return messages instead of plain text
)

# 3. Set up the conversation chain with the summarizing memory
conversation = ConversationChain(
    llm=llm,
    memory=summary_memory,
    verbose=True
)

# 4. Use the conversation
response = conversation.predict(input="Hi, I need help hiring a Python developer in Bangalore.")
print("response 1:",response)

response = conversation.predict(input="They should have experience with Flask and 2+ years of backend development.")
print("response 2:",response)

# Optional: check what's in memory
print("memory:::::")
print(summary_memory.buffer)  # shows the summary
```

day3 assigmnent:

You are hired by a startup called CareConnect, which is building an AI-powered assistant to support virtual healthcare providers. This assistant will help nurses, doctors, and patients by analyzing chat transcripts of telehealth sessions.

Your job is to design LangChain tools and build an intelligent agent that can dynamically invoke the right tool based on what the user is asking.

 **Problem Statement**:
Given a doctor-patient conversation, the AI assistant should help by:


*   Detecting the patient's emotional state. Give a score on a scale of 1-5.
*   Summarizing the chief complaint and urgency of the patient.


```python
!pip install langchain-groq
```

```python


from langchain.agents import initialize_agent, AgentType
from langchain.tools import tool
from langchain_groq import ChatGroq
import os




llm = ChatGroq(
    groq_api_key=os.environ["GROQ_API_KEY"],
    model_name="llama3-70b-8192"
)

@tool
def detect_emotion_state(conversation: str) -> str:
    """Detect the patient's emotional state on a scale of 1 (calm) to 5 (very distressed)."""
    return f"""Given the conversation below, rate the patient's emotional state from 1 (calm) to 5 (very distressed).

Conversation:
{conversation}

Format:
Emotional State: <1-5>
Reason: <short explanation>
"""

@tool
def summarize_chief_complaint(conversation: str) -> str:
    """Summarize the patient's chief complaint and rate urgency (low, medium, high)."""
    return f"""Summarize the patient's chief complaint and assess urgency (low, medium, high).

Conversation:
{conversation}

Format:
Chief Complaint: <summary>
Urgency: <low/medium/high>
"""


tools = [detect_emotion_state, summarize_chief_complaint]
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

conversation = """
Doctor: Hello! What brings you in today?
Patient: I’ve been having really bad headaches and dizziness for the past three days.
Doctor: Have you taken anything for it?
Patient: Just paracetamol, but it’s not helping. I’m starting to get really worried.
"""


response = agent.run(f"Analyze the following conversation:\n{conversation}")
print(response)
```
